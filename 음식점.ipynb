```python
import pyproj
import pandas as pd
import numpy as np
```

```python
seoul_food_loc = pd.read_csv("서울시 일반음식점 위치 정보.csv", thousands = ',',
                                encoding = 'utf-8')
seoul_food_loc
```

```python
def project_array(coord, p1_type, p2_type):
    """
    좌표계 변환 함수
    - coord: x, y 좌표 정보가 담긴 NumPy Array
    - p1_type: 입력 좌표계 정보 ex) epsg:5179
    - p2_type: 출력 좌표계 정보 ex) epsg:4326
    """
    p1 = pyproj.Proj(init=p1_type)
    p2 = pyproj.Proj(init=p2_type)
    fx, fy = pyproj.transform(p1, p2, coord[:, 0], coord[:, 1])
    return np.dstack([fx, fy])[0]
```

```python
coord = np.array(seoul_food_loc)
coord
```

```python
p1_type = "epsg:2097"
p2_type = "epsg:4326"

result = project_array(coord, p1_type, p2_type)
result
```

```python
seoul_food_loc['경도'] = result[:, 0]
seoul_food_loc['위도'] = result[:, 1]
```

```python
seoul_food_loc
```

```python
seoul_food = pd.read_csv("서울시 일반음식점 정보.csv", thousands = ',',
                                encoding = 'utf-8')
seoul_food
```

```python
seoul_food['경도'] = seoul_food_loc['경도']
seoul_food['위도'] = seoul_food_loc['위도']
seoul_food
```
