```python
import pyproj
```

```python
# TM 좌표계
tm = pyproj.CRS.from_proj4("+proj=tmerc +lat_0=38N +lon_0=127.5E +ellps=GRS80 +units=m +no_defs")

# 위경도 좌표계
wgs84 = pyproj.CRS.from_epsg(4326)  # WGS84 EPSG code

# TM 좌표
x = 961607.767162
y = 1954006.611657

# TM 좌표 -> 위경도 좌표 변환
transformer = pyproj.Transformer.from_crs(tm, wgs84, always_xy=True)
lon, lat = transformer.transform(x, y)

print(lon, lat)
```
