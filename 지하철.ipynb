```python
seoul_subway_lat = [] #위도
seoul_subway_lng = [] #경도

for x in seoul_subway['도로명주소']:
    tmp = gmaps.geocode(x, language = 'ko')
    tmp_loc= tmp[0].get("geometry")

    seoul_subway_lat.append(tmp_loc['location']['lat'])
    seoul_subway_lng.append(tmp_loc['location']['lng'])
```
